```julia
# load the JuMP package and Cbc solver package
using JuMP, Cbc

# create the model object
m = Model()

@variable(m, x[1:5,1:5], Bin)
@variable(m, w[1:5,1:5], Bin)
@variable(m, y[1:5,1:5], Bin)
@variable(m, z[1:5,1:5], Bin)
@variable(m, p[1:5,1:5], Bin)

@objective(m, Max, sum(x + w + y + z + p))

@constraint(m, const_1, x[1,1] == 0)
@constraint(m, const_2, w[3,3] == 0)
@constraint(m, const_2_2, y[4,4] == 1)
@constraint(m, const_2_3, z[4,5] == 0)
@constraint(m, const_3, z[1,3] + z[2,3] == 0)
@constraint(m, const_3_2, z[3,3] == 0)
@constraint(m, const_3_3, y[3,2] == 0)
@constraint(m, const_4, (x[2,1] + x[2,2]) + (x[4,1] + x[4,2]) + (x[5,1] + x[5,2]) == 0)
@constraint(m, const_4_2, w[5,1] == 0)
@constraint(m, const_4_3, y[5,1] == 1)
@constraint(m, const_4_4, z[5,1] == 1)
@constraint(m, const_5, w[4,4] + w[4,5] == 1)
@constraint(m, const_5_2, p[4,4] + p[4,5] == 1)
@constraint(m, const_6, p[3,5] == 1)
@constraint(m, x_assign[i in 1:5], sum(x[i,j] for j in 1:5) == 1)
@constraint(m, w_assign[i in 1:5], sum(w[i,j] for j in 1:5) == 1)
@constraint(m, y_assign[i in 1:5], sum(y[i,j] for j in 1:5) == 1)
@constraint(m, z_assign[i in 1:5], sum(z[i,j] for j in 1:5) == 1)
@constraint(m, p_assign[i in 1:5], sum(p[i,j] for j in 1:5) == 1)

# specify the solver you want to use to solve Model m
set_optimizer(m, Cbc.Optimizer)

# solve it
optimize!(m)

# display the solution
solution_summary(m,verbose=true)
```
