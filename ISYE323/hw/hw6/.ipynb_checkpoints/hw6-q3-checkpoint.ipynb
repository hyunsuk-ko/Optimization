```julia
using JuMP, Cbc

m = Model()

@variable(m, x_s1 >= 0)
@variable(m, x_s2 >= 0)
@variable(m, x_s3 >= 0)
@variable(m, x_11 >= 0)
@variable(m, x_12 >= 0)
@variable(m, x_21 >= 0)
@variable(m, x_22 >= 0)
@variable(m, x_23 >= 0)
@variable(m, x_24 >= 0)
@variable(m, x_31 >= 0)
@variable(m, x_32 >= 0)
@variable(m, x_33 >= 0)
@variable(m, x_1t >= 0)
@variable(m, x_2t >= 0)
@variable(m, x_3t >= 0)
@variable(m, x_4t >= 0)
@variable(m, x_ts >= 0)

@objective(m, Max, x_ts)

@constraint(m, max_playtime_1, x_s1 <= 15)
@constraint(m, max_playtime_2, x_s2 <= 12)
@constraint(m, max_playtime_3, x_s3 <= 13)
@constraint(m, flowbalance_camp1, x_s1 == x_11 + x_12)
@constraint(m, flowbalance_camp2, x_s2 == x_21 + x_22 + x_23 + x_24)
@constraint(m, flowbalance_camp3, x_s3 == x_31 + x_32 + x_33)
@constraint(m, flowbalance_month1, x_11 + x_21 + x_31 == x_1t)
@constraint(m, flowbalance_month2, x_12 + x_22 + x_32 == x_2t)
@constraint(m, flowbalance_month3, x_23 + x_33 == x_3t)
@constraint(m, flowbalance_month4, x_24 == x_4t)
@constraint(m, max_monthly_time1, x_1t <= 10)
@constraint(m, max_monthly_time2, x_2t <= 10)
@constraint(m, max_monthly_time3, x_3t <= 10)
@constraint(m, max_monthly_time4, x_4t <= 10)
@constraint(m, max_campaign1_month1, x_11 <= 8)
@constraint(m, max_campaign2_month1, x_21 <= 8)
@constraint(m, max_campaign3_month1, x_31 <= 8)
@constraint(m, max_campaign1_month2, x_12 <= 8)
@constraint(m, max_campaign2_month2, x_22 <= 8)
@constraint(m, max_campaign3_month2, x_22 <= 8)
@constraint(m, max_campaign2_month3, x_23 <= 8)
@constraint(m, max_campaign3_month3, x_33 <= 8)
@constraint(m, max_campaign2_month4, x_24 <= 8)
@constraint(m, flowbalance_st, x_s1 + x_s2 + x_s3 == x_ts)

set_optimizer(m, Cbc.Optimizer)

optimize!(m)

solution_summary(m, verbose=true)
```
