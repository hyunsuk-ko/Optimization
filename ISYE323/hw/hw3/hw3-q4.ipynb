```julia
# 20 mining sites, just numbered as 1...20
mines = 1:20

# 7 houses, just numbered as 1..7
houses = 1:7

# cost of spice production at each mine (in Solari)
mine_prod_cost =  [19, 27, 16, 17, 14, 18, 8, 14, 6, 21, 17, 14, 19, 15, 12, 6, 8, 28, 12, 21]

# amount of spice at each mine (tons) 
# this value is a constant for all mines
mine_cap = 50

# demand for spice (tons) at each House
demand = [70, 61, 72, 57, 45, 81, 79] 

# rows are mines, columns are Houses
transport_cost = [
2.79 1.21 1.48 0.59 1.70 1.84 1.94;
2.09 0.26 0.26 0.27 1.38 1.60 0.27;
0.56 2.89 2.72 1.28 1.66 0.02 1.39; 
2.96 1.02 0.13 2.33 0.78 0.69 0.17; 
1.09 2.06 0.72 2.82 2.60 0.94 0.22; 
1.19 2.91 0.54 2.56 1.73 2.98 0.43; 
1.94 0.94 2.47 0.82 0.19 1.91 0.42; 
2.15 1.42 0.30 0.33 1.35 1.18 0.13; 
2.96 1.95 1.00 0.46 2.14 0.38 1.88; 
1.48 2.53 1.53 2.57 2.54 2.29 2.10; 
2.65 2.11 0.53 1.36 1.16 0.44 1.61; 
2.12 1.88 0.62 1.22 1.40 2.37 1.48; 
1.45 2.32 2.89 1.74 1.31 1.30 0.13; 
2.68 1.25 1.32 0.82 2.90 0.51 1.55; 
1.75 0.40 0.21 0.24 1.68 0.78 0.67; 
2.00 2.52 1.07 2.98 0.22 0.08 1.46; 
0.61 2.36 1.11 0.54 2.77 0.85 1.79; 
0.63 0.63 0.29 1.30 2.45 1.74 0.36; 
1.14 2.76 0.34 1.16 0.13 1.74 0.95; 
1.00 1.30 0.29 1.48 1.81 0.12 1.07]

# percent of sand in spice at each mine
sand_per = 
[7.0, 5.7, 3.5, 1.3, 8.9, 
 1.0, 8.7, 5.4, 5.0, 1.4, 
 3.8, 5.3, 8.0, 7.4, 5.0,
 2.2, 7.8, 1.9, 4.1, 7.7]

# percent of sulfur in spice at each mine
sulf_per = 
[3.8, 4.1, 4.2, 7.4, 8.9,
 5.6, 1.5, 1.6, 7.1, 2.0,
 5.5, 0.6, 0.8, 2.0, 8.0,
 4.4, 7.2, 8.6, 7.0, 8.1]

# max percent of sand at each House
sand_lim = 6
# max percent of sulfur at each House
sulf_lim = 7
;
```

```julia
using JuMP, Cbc

m = Model()

@variable(m, x[mines, houses] >= 0)

@objective(m, Min, sum(mine_prod_cost[i] * x[i,j] for i in mines for j in houses) + sum(transport_cost[i,j] * x[i,j] for i in mines for j in houses))

@constraint(m, demand_req[j in houses], sum(x[i,j] for i in mines) >= demand[j])
@constraint(m, supply_req[i in mines], sum(x[i,j] for j in houses) <= mine_cap)
@constraint(m, sand_req[j in houses], sum(sand_per[i] * x[i,j] for i in mines) <= demand[j] * sand_lim)
@constraint(m, sulf_req[j in houses], sum(sulf_per[i] * x[i,j] for i in mines) <= demand[j] * sulf_lim)

set_optimizer(m, Cbc.Optimizer)

optimize!(m)

solution_summary(m, verbose=true)
```
