```julia
using Cbc, JuMP, NamedArrays

# Sets
T = [1, 2, 3, 4, 5, 6, 7]
T0 = [0, 1, 2, 3, 4, 5, 6, 7]
L = [1, 2]

# Parameters
c = 0.5
s = 0.2
b = Dict(zip(L, [150, 260]))
d = NamedArray([100 130 150 120 140 220 150; 150 230 200 190 280 140 320], (L, T), ("location", "day"))
h = 0.1

########################################
# TODO: create and solve general model #
########################################

m = Model()

@variable(m, 0 <= x[l in L, t in T] <= b[l])
@variable(m, y[L, T0] >= 0)
@variable(m, z[L, T0] >= 0)

@objective(m, Min, sum(sum(c*x[l, t] + s*y[l, t] + h*z[l, t] for l in L) for t in T))

@constraint(m, storeandship1[t in T], y[1,t] == y[1,t-1] + x[1,t] + z[2,t-1] - z[1,t-1] - d[1,t])
@constraint(m, storeandship2[t in T], y[2,t] == y[2,t-1] + x[2,t] + z[1,t-1] - z[2,t-1] - d[2,t])
@constraint(m, shipbound[l in L, t in T], z[l, t] <= y[l,t])
@constraint(m, startingstock[l in L], y[l, 0] == 0)
@constraint(m, startingship[l in L], z[l, 0] == 0)

set_optimizer(m, Cbc.Optimizer)

optimize!(m)

solution_summary(m, verbose=true)
```
