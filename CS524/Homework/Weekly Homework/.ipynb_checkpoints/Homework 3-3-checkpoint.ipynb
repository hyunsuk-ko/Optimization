## Problem 1 -- Putting together a team

Nick Fury needs help creating a team of superheros to help him protect the earth. He has a set of candidates and a set of skills he'd like his team to have. He's categorized his possible team members into four groups: \{Government affiliates, Superhumans, Guardians of the Galaxy, Other\}. Each candidate belongs to exactly one of the four groups. Each candidate also exhibits some number of the skills in his list. Fury wants to pull no more than 4 members from each group, but he also wants to ensure that all the skills are "covered"; in other words, each skill must belong to at least one member of the final team.

(a) Write a __maximum flow problem__ that will tell Fury whether
or not there is a feasible assignment of candidates to  the team that will meet his requirements. _Hint:_  Draw the network first. _Hint2.0:_ There is some code below that will help you get started building the nodes and arcs in Julia. There is more than one way to build this network; you do not have to follow the structure I have given below.

```julia
# types of beings 
classes = [:government, :superhumans, :guardians, :other] 

# all the possible candidates for the team
candidates = [:clint_barton, :daisy_johnson,:nick_fury, :phil_coulson, :maria_hill, :james_rhodes, :natasha_romanoff,
    :tony_stark,:bruce_banner, :steve_rogers, :wanda_maximioff, :tchalla, :sam_wilson, :peter_parker, :scott_lang,
    :luke_cage, :danny_rand, :stephen_strange, :matt_murdock,  :hope_vandyne, :okoye, :bucky_barnes,
    :mantis, :yondu,  :rocket, :nebula, :peter_quill, :gamora,:drax, :groot,
    :thor, :vision, :carol_danvers,:valkyrie,:loki, :talos, :pepper_potts, :jane_foster, :darcy_lewis, :erik_selvig]

# candidates that are part of the government
govt = [:clint_barton, :daisy_johnson,:nick_fury, :phil_coulson, :maria_hill, :james_rhodes, :natasha_romanoff]

# unaffiliated earth-based superhumans
human = [:tony_stark,:bruce_banner, :steve_rogers, :wanda_maximioff, :tchalla, :sam_wilson, :peter_parker, :scott_lang,
    :luke_cage, :danny_rand, :stephen_strange, :matt_murdock, :jessica_jones,  :hope_vandyne, :okoye, :bucky_barnes ]

# candidates that are affiliated with the guardians of the galaxy
guardian = [:mantis, :yondu,  :rocket, :nebula, :peter_quill, :gamora,:drax, :groot]

# candidates that are classified as "other"
other = [:thor, :vision, :carol_danvers, :valkyrie, :loki, :talos, :pepper_potts, :jane_foster, 
    :heimdall, :darcy_lewis, :erik_selvig]

# skills that are needed on the team
skill_areas = [:flying, :super_strength, :indestructibility, :intelligence, :wisdom, 
    :hacking, :leadership, :accuracy, :stealth, :illusions, :engineering, :grit, :time_travel,
    :matter_manipulation, :charisma, :intimidation]



using Random

# set a seed so we get the same random numbers every time
seed = 547354334
Random.seed!(seed)

# randomly match candidates to skill areas
skill_candidate = Dict() 
density = 0.1 # 10% of attempted matches will be successful (lower density -> lower success rate)

for c in candidates
    for s in skill_areas
        # if a uniform(0,1) number is less than density
        if Random.rand() < density
            # the candidate is an expert in that area
            skill_candidate[(s,c)] = 1
        end
    end
end

# we need at least one candidate per skill area
num_from_skill = Dict(zip(skill_areas,ones(length(skill_areas))))

# we can have at most 4 individuals from each class of beings
max_from_class= Dict(zip(classes,4*ones(length(classes))));

# create the list of nodes in the graph 
nodes = vcat([:source], skill_areas, candidates, classes, [:sink]);

# create all arcs needed in the network
arcs_1 = [(:source,i) for i in skill_areas] # arcs from source to sklls;
arcs_2 = [n[1] for n in skill_candidate] # arcs from skills to experts in that skill
arcs_govt= []
arcs_human= []
arcs_guardian= []
arcs_other= []

for candidate in candidates
    if candidate in govt
        push!(arcs_govt, (candidate,:government))
    elseif candidate in human
        push!(arcs_human, (candidate,:superhumans))
    elseif candidate in guardian
        push!(arcs_guardian, (candidate,:guardians))
    else
        push!(arcs_other, (candidate,:other))
    end
end

arcs_to_sink = [(class, :sink) for class in classes]
arcs_dummy = [(:sink,:source)]
    

### CREATE THE REST OF THE ARCS HERE ###

arcs = vcat(arcs_1, arcs_2); # concatenate all lists of arcs together in one master arc list
arcs_group = vcat(arcs_govt, arcs_human, arcs_guardian, arcs_other)
arcs = vcat(arcs, arcs_group)
arcs = vcat(arcs, arcs_to_sink)
arcs = vcat(arcs, arcs_dummy)
```

```julia
### DATA ###

# Capacity: 1 for normal arcs / 4 for group arcs / inifinite for dummy arc
capacity = Dict()
for arc in arcs
    if arc[1] == :source || arc[1] in skill_areas  || arc[1] in candidates
        capacity[arc] = 1
    elseif arc[1] == :superhumans || arc[1] == :government || arc[1] == :guardians || arc[1] == :other
        capacity[arc] = 4
    else
        capacity[arc] = 1000000
    end
end

c = [] # Costs: should be 0 except for sink (-1)
for i in 1:125
    if i != 125
        push!(c, 0)
    else
        push!(c, -1)
    end
end
c = Dict(zip(arcs, c))


# Incidence Matrix with nodes on rows and arcs on columns
using NamedArrays 
result = NamedArray(zeros(62,125),(nodes,arcs), ("Nodes", "Arcs"))

for node in nodes
    for arc in arcs 
        start = arc[1]
        finish = arc[2]
        if arc[1] == node
            result[start,arc] = 1 # supply
            result[finish,(start,finish)] = -1 # demand
        end
    end
end

A = Matrix(result)

# supply & demand both should be 0 on max flow problem
b = []
for i in 1:62
    push!(b, 0.0)
end


### MODEL ###
using JuMP, Clp

m = Model(Clp.Optimizer)

@variable(m, x[arc in arcs] >= 0)

@objective(m, Min, sum(c[arc] * x[arc] for arc in arcs))

# don't exceed arc capacity
@constraint(m, arc_cap[arc in arcs], x[arc] <= capacity[arc])

# No more than 4 members from each group
@constraint(m, group_req[class in classes], x[(class,:sink)] <= 4)

# constraints balance flow into and out of each node
@constraint(m, balance_flow[node in nodes], sum(result[node,arc] * x[arc] for arc in arcs) == 0)

# All skills should be covered
#@constraint(m, skills_covered[skill in skill_areas], )

optimize!(m)

println("Max flow: ", -objective_value(m))

for arc in arcs
    println("The flow from ", arc[1], " to ", arc[2], " is ", value(x[arc]))
end
```

## Problem 3 -- A graphical approach to duality

Consider the following linear program:

\begin{align}
\max & 3x_1 + x_2\\
s.t. & 4x_1 + x_2 \leq 7\\
& 5x_1 + 2x_2 \leq 10\\
& x_1, x_2 \geq 0
\end{align}

(a) Write the dual of this linear program.

\begin{align}
\min &  7λ_1 + 10λ_2 \\
s.t. & 4λ_1 + 5λ_2 \geq 3 \\
& λ_1 + 2λ_2 \geq 1 \\
& λ_1  \geq 0,  λ_2 \geq 0 \\
\end{align}

(b) Is an objective value of 17/3, $x_1 = 4/3$, and $x_2 = 5/3$ an optimal solution to this problem? Justify your answer with appropriate theory.

 We know that $x_1* = 4/3$, $x_2* = 5/3$,  and the solution of the primal is 17/3. Also, according to strong duality, p* = d* = 17 / 3. 

  In order to minimize the optimal dual solution, dual variables should be less as possible, therefore below equations hold:
\begin{align}
\ & 4λ_1 + 5λ_2 = 3 \\
& λ_1 + 2λ_2 = 1 \\
\end{align}

From above equations, we can derive $λ_1* = 1/3$, $λ_2* = 1/3$. Consequently, the optimal solution of the dual is $7λ_1 + 10λ_2 = 17 / 3$.

Therefore, $p* = d* = 17/3$.
